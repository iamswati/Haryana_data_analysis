# Haryana - EDA
### This NoteBook starts with exploratory data analysis for the district Haryana, India. The data contains various features regarding the COVID-19 outbreak like Date, TotalSamples, etc

```python
# Importing required libraries

# Matplotlib is a low level graph plotting library in python that serves as a visualization utility
import matplotlib.pyplot as plt
import matplotlib as plt
import numpy as np          # linear algebra      
import pandas as pd         # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Reading CSV File
df=pd.read_csv("Haryana covid-19.csv")
df
```

```python
# View shape of data - rows and columns
df.shape
```

```python
# describe the data
df.describe()
```

```python
# Getting data-type
df.info()
```

```python
# View columns
df.columns
```

```python
# Convert the Integers to Floats Data Type

df['TotalSamples'] = df['TotalSamples'].astype(float)
df['Negative'] = df['Negative'].astype(float)
```

```python
# Getting data-type
df.info()
```

```python
#counting number of null values
df.isnull().sum()
```

```python
# split date 
date=df[["Date"]]
group_date=date.groupby("Date")
group_date.size()
```

```python
df["Date"] =pd.to_datetime(df["Date"])
df["Date"]
```

```python
df["Date"].dt.year
```

```python
df["Date"].dt.month
```

```python
df["Date"].dt.day
```

```python
# New columns are created- year, month, and day
df["year"] = df["Date"].dt.year
df["month"] = df["Date"].dt.month
df["day"] = df["Date"].dt.day
df
```

### Total Number of samples, negative cases and positive cases

```python
tnocr=df["TotalSamples"].sum(axis = 0, skipna = True)
print("Total Number of samples:", tnocr)
```

```python
tnocr=df["Negative"].sum(axis = 0, skipna = True)
print("Total Number of negative cases:", tnocr)
```

```python
tnocr=df["Positive"].sum(axis = 0, skipna = True)
print("Total Number of positive cases:", tnocr)
```

```python
sns.histplot(x=df.TotalSamples, kde=True).set_title("Plot Distribution of TotalSamples")
```

```python
sns.histplot(x=df.Negative, kde=True).set_title("Plot Distribution of Negative")
```

```python
sns.histplot(x=df.Positive, kde=True).set_title("Plot Distribution of Positive")
```

### Plot line graph- TotalSamples, Positive cases and Negative cases with respect to Date

```python
sns.lineplot(x=df.Date,y='TotalSamples',data=df).set_title("TotalSamples vs date for Haryana")
```

```python
sns.lineplot(x=df.Date,y='Negative',data=df).set_title("Negative cases vs date for Haryana")
```

```python
sns.lineplot(x=df.Date,y='Positive',data=df).set_title("Positive cases vs date for Haryana")
```

```python
sns.lineplot(x=df.Negative,y='Positive',data=df).set_title("Positive cases vs Negative cases for Haryana")
```

```python
df_corr = df[['Date', 'TotalSamples', 'Negative', 'Positive']].corr()
df_corr
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.heatmap(df_corr, vmin=0, vmax=1,annot=True)
```
